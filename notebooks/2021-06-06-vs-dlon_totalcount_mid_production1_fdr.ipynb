```r
# Data Import from ../../models/ 
```

```r
final_results<- read.table("../../models/TC_mid_production1_2_3_4_5_regularization.tsv",header=T)
tnmean <- read.table("../../data/processed/TC_mid_production1_2_3_4_5_mean.tsv",header=T)
```

```r
##FDR
```

```r
path <- ("../../reports/production1_2_3_4_5/mid/TC_fdr_csv/")
```

```r
dflist <- rnbtn_fdr_effects(final_results,tnmean,locus_tag = "locus_tag",
  exeffects = c("Intercept", "batch", "log"),
  fecutoff = 1,
  cecutoff = 1,method="fdrtool")
```

```r
essential_df <- dflist[[1]]
modelessential_df <- dflist[[2]]
library(stringr)
```

```r
lapply(1:length(essential_df), function(i) write.csv(essential_df[[i]], file.path(path,paste0(str_replace_all(names(essential_df[i]),":","_"), "_essential.csv")),
                                      row.names = FALSE))
```

```r
lapply(1:length(modelessential_df), function(i) write.csv(modelessential_df[[i]], file.path(path,paste0(str_replace_all(names(modelessential_df[i]),":","_"), "_effect.csv")),
                                      row.names = FALSE))
```
