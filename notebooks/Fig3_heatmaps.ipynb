```r
split <- c(rep('WT',20),rep('DCLPA',39),rep('DCLPB',44),rep('DLON',33),rep('dnaKJ-NI',38))
```

```r
rownames(mat) <- gsub(".*_","",rownames(mat))
colnames(mat) <- c("CN:H","CN:L","CN:M","HT:H","HT:L","HT:M","OS:H","OS:L","OS:M")
order.col <- c("CN:L","CN:M","CN:H","HT:L","HT:M","HT:H","OS:L","OS:M","OS:H")
mat <- mat[,order.col]
```

```r
ddf <- data.frame(genes=row.names(mat),strain=split)
ddgenes <- (ddf%>%get_dupes(genes))$genes
```

```r
options(repr.plot.width=12, repr.plot.height=10)
pdf(file="../../PNAS/FIGURES/Fig3c_2.pdf",width=13,height=11)
circos.par(start.degree = 130, gap.degree = 10)
circos.heatmap(mat, split = split, col = col_fun, track.height = 0.40,cluster = TRUE,
    bg.border = "grey", bg.lwd = 2, bg.lty = 2, show.sector.labels = TRUE, rownames.side = "inside"
)   
circos.track(track.index = get.current.track.index(), panel.fun = function(x, y) {
    if(CELL_META$sector.numeric.index == 1|CELL_META$sector.numeric.index == 2|CELL_META$sector.numeric.index == 3|CELL_META$sector.numeric.index == 4|CELL_META$sector.numeric.index == 5) { # the last sector
        cn = colnames(mat)
        n = length(cn)
        circos.text(rep(CELL_META$cell.xlim[2], n) + convert_x(1, "mm"), 
            n - 1:n + 0.5, cn, 
            cex = 0.5, adj = c(0, 0.5), facing = "inside")
    }
}, bg.border = NA)
for(i in seq_len(nrow(matrixdf))) {
    circos.heatmap.link(matrixdf$V1[i],
                        matrixdf$V2[i],
                        col = "grey")
}
lgd = Legend(title = "fitness", col_fun = col_fun)
draw(lgd,just = c("right", "bottom"))
dev.off()
```
