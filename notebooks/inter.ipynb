```r
map(2:length(l),
      ~ combn(l, .x, \(x)
              list(reduce(x, intersect)) %>%
                set_names(str_c(names(
                  x
                ), collapse = ' & ')),
              simplify = FALSE)) %>%
  unlist(FALSE) %>%
  unlist(FALSE) %>%
  c(.,
    map(seq_along(l), ~reduce(l[-.x], setdiff,.init = l[[.x]])) %>% 
      set_names(names(l))
    ) %>% 
  keep(~ length(.x) > 0)
```
